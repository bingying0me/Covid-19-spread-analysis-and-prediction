```python
import tensorflow as tf
import pandas as pd
```

```python
data = pd.read_csv("./dataset/EverydayCase.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
tick_spacing = 4
%matplotlib inline
plt.figure(figsize=(15, 5))
plt.title("The number of the Confirmed Case in Hong Kong (Everyday)", fontsize=20)
plt.xlabel("Date", color='r')
plt.ylabel("Number of people", color='r')
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.plot(data.date, data.confirmedcases)
plt.show()
```
